```python
from scipy.stats import pearsonr
import pandas as pd

combined_df = pd.read_csv('Combined_Sentiment_and_Performance_Data.csv')

combined_df.head()

sentiment_wins_corr, _ = pearsonr(combined_df["Sentiment"], combined_df["recent_wins"])
sentiment_losses_corr, _ = pearsonr(combined_df["Sentiment"], combined_df["recent_losses"])

sentiment_wins_corr, sentiment_losses_corr
```

```python
import matplotlib.pyplot as plt
combined_df['Win_Rank'] = combined_df['recent_wins'].rank(ascending=False, method='min')

combined_df['Sentiment_Rank'] = combined_df['Sentiment'].rank(ascending=False, method='min')

combined_df_sorted = combined_df.sort_values(by="Win_Rank")

plt.figure(figsize=(12, 6))
plt.bar(combined_df_sorted['Team'], combined_df_sorted['Win_Rank'], label="Win Rank")
plt.plot(combined_df_sorted['Team'], combined_df_sorted['Sentiment_Rank'], color="red", marker='o', linestyle='-', label="Sentiment Rank")

plt.xlabel("Teams (Sorted by Win Rank)")
plt.ylabel("Rank")
plt.title("Team Win Rank vs Sentiment Rank")
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

win_rank_mean = combined_df['Win_Rank'].mean()
sentiment_rank_mean = combined_df['Sentiment_Rank'].mean()

win_rank_median = combined_df['Win_Rank'].median()
sentiment_rank_median = combined_df['Sentiment_Rank'].median()

win_rank_std = combined_df['Win_Rank'].std()
sentiment_rank_std = combined_df['Sentiment_Rank'].std()

win_rank_min = combined_df['Win_Rank'].min()
sentiment_rank_min = combined_df['Sentiment_Rank'].min()

win_rank_max = combined_df['Win_Rank'].max()
sentiment_rank_max = combined_df['Sentiment_Rank'].max()

correlation = combined_df['Win_Rank'].corr(combined_df['Sentiment_Rank'])

print("Statistical Analysis:")
print(f"Win_Rank - Mean: {win_rank_mean:.2f}, Median: {win_rank_median:.2f}, Std Dev: {win_rank_std:.2f}, Min: {win_rank_min}, Max: {win_rank_max}")
print(f"Sentiment_Rank - Mean: {sentiment_rank_mean:.2f}, Median: {sentiment_rank_median:.2f}, Std Dev: {sentiment_rank_std:.2f}, Min: {sentiment_rank_min}, Max: {sentiment_rank_max}")

print("\nCorrelation Analysis:")
print(f"Correlation between Win_Rank and Sentiment_Rank: {correlation:.2f}")
```
